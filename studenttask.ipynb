```python
#4)Find the total and average of the exam, quiz and homework and store them in a separate collection.
stage1 = {"$unwind":"$scores"}
stage2 = {"$group":{"_id":"$scores.type","Total":{"$sum":"$scores.score"},"Avg_score":{"$avg":"$scores.score"}}}
stage3 = {"$out":"Total_Average"}
for i in newcol.aggregate([stage1,stage2,stage3]):
    pprint(i)
x = newdb.list_collection_names()
print(x)
for i in newdb.Total_Average.find():
    print(i)
```

```python
#6)Create a new collection which consists of students who scored below the fail mark in all the categories.
stage1 = {"$unwind":"$scores"}
stage2 = {"$project":{"_id":0,"name":1,"fail":{"$sum":"$scores.score"}}}
stage3 = {"$match":{"fail":{"$lt":40}}}
stage4 = {"$out":"Fail_stu_col"}
for i in newcol.aggregate([stage1,stage2,stage3,stage4]):
    print(i)
newdb.list_collection_names()
for i in newdb.Fail_stu_col.find():
    print(i)
```
